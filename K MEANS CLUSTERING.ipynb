# IMPLEMEMNTING K MEANS ALGORITHM 

IMPORT THE REQUIRED LIBARARIES

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

LOAD AND READ THE DATA

```python
df = pd.read_csv("C:\MACHINE LEARNING\EDA AND FEATURE ENGINEERING\ENDURE WORK\demos\DATA\Mall_Customers\Mall_Customers.csv")
```

```python
X_data = df[["Annual Income (k$)", "Spending Score (1-100)"]]
```

```python
df.head()
```

EXPLORING THE DATA

```python
df.columns
```

```python
df.isnull().sum()
```

SCALING THE DATA

```python
kmeans_scaler=StandardScaler()
x_scaled=kmeans_scaler.fit_transform(X_data)
```

USING KMEANS FROM SKLEARN AND FIT_PREDICT TO ASSIGN CLUSTER LABELS

```python
kmeans_clustering = KMeans(n_clusters=5, random_state=42, n_init=5)
y_kmeans = kmeans_clustering.fit_predict(x_scaled)
```

EVALUATION USING INERTIA AND SILHOUTTE SCORE

```python
print("Inertia (WCSS):", kmeans_clustering.inertia_)
print("Silhouette Score:", silhouette_score(x_scaled, y_kmeans))
```

VISUALIZATION

```python
plt.scatter(x_scaled[:, 0], x_scaled[:, 1], c=y_kmeans, cmap="viridis", s=50)
plt.scatter(kmeans_clustering.cluster_centers_[:, 0], kmeans_clustering.cluster_centers_[:, 1],
            c="red", s=200, marker="X", label="Centroids")
plt.xlabel("Annual Income (scaled)")
plt.ylabel("Spending Score (scaled)")
plt.title("K-Means Clustering - Mall Customers")
plt.legend()
plt.show()
```
